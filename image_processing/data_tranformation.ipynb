## Renaming images in the folder

+ rename the filename of the image
+ check if images has other extensions: jpeg, png, jpg, etc. 

```python
# check the extensions of images 
extensions = ('.png', '.jpg', '.jpeg', '.bmp', '.gif')
png, jpg, jpeg, bmp, gif = np.zeros(len(extensions))
```

```python
# Assuming images_list is defined
for image_file in image_files:
    if image_file.endswith(extensions[0]):
        png += 1
    elif image_file.endswith(extensions[1]):
        jpg += 1
    elif image_file.endswith(extensions[2]):
        jpeg += 1
    elif image_file.endswith(extensions[3]):
        bmp += 1
    elif image_file.endswith(extensions[4]):
        gif += 1

# Display results
print("PNG files:", png)
print("JPG files:", jpg)
print("JPEG files:", jpeg)
print("BMP files:", bmp)
print("GIF files:", gif)

```

```python
new_names = [f"smartphone_{count}.jpg" for count in range(len(image_files))]

# Loop through the image files and rename them
for old_name, new_name in zip(image_files, new_names):
    old_path = os.path.join(img_dir, old_name)
    new_path = os.path.join(img_dir, new_name)

    # Check if the new file already exists
    if not os.path.exists(new_path):    
        os.rename(old_path, new_path)  # Rename file
        print(f"Renamed {old_name} to {new_name}")
    else:
        print(f"Skipping renaming: {new_name} already exists")
```
