## Data Transformation

```python
# combine image file name with its path
image_path = "D:\\Machine_Learning\\Portfolio_Project_Machine_Learning\\Mobile_Phone_Recognition\\mobile_phone_images\\mobile_images"
images_ar = np.array([os.path.join(image_path, image_file) for image_file in images_list])

# display shape of images and labels
print(f"Images: {images_ar.shape}")
print(f"Labels: {label_ar.shape}")
```

```python
# implement function 4: construct a column list
def construct_dataframe(features: list, data: list) -> dict: 
  web_content = {feature: sample for feature, sample in zip(features, data)}
  dataframe = pd.DataFrame(web_content)
  return dataframe

# convert image urls and labels into dataframe
graph_data = [images_ar, label_ar]
feature_name = ["Image_path", "Mobile_label"]
image_df = construct_dataframe(features=feature_name, data=graph_data)
```

## Data Cleaning

Preprocessing the mobile label data based on these tasks/ 

+ remove special characters 
+ extract only brand labels

```python
# retrieve mobile label data
image_df["Mobile_label"] = image_df["Mobile_label"].apply(lambda text: text.lower())
image_df["Mobile_brand"] = image_df["Mobile_label"].str.split(" ", expand=True)[0]
```
